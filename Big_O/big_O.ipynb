## Big O Notation 

Big O notation is a mathematical notation used in computer science to describe the efficiency or complexity of an algorithm concerning the size of its input. It provides a way to express how the runtime or space requirements of an algorithm grow as the input size increases. Big O notation is crucial for analyzing and comparing the performance of algorithms without getting bogged down by specific implementation details.

### Key points about Big O notation:

Scaling Efficiency: Big O notation characterizes the worst-case behavior of an algorithm as the input size approaches infinity. It helps in understanding how the algorithm's performance scales with larger inputs.
Simplification: Big O notation simplifies the analysis by focusing on the most significant factors affecting an algorithm's performance and ignoring constant factors and lower-order terms.

Common Notations:
* O(1): Constant time complexity (e.g., accessing an element in an array).
* O(log n): Logarithmic time complexity (e.g., binary search).
* O(n): Linear time complexity (e.g., iterating through an array).
* O(n log n): Log-linear time complexity (e.g., efficient sorting algorithms like merge sort and heap sort).
* O(n^2): Quadratic time complexity (e.g., nested loops).
* O(2^n): Exponential time complexity (e.g., recursive algorithms without memoization).

Space Complexity: Big O notation is also used to analyze the space (memory) efficiency of an algorithm.
Comparative Analysis: It allows for comparing algorithms and selecting the most efficient one for a specific task, considering scalability and resource constraints.
Focus on Dominant Terms: When expressing the time complexity with Big O notation, only the most dominant term or the term with the highest growth rate is considered. This provides a high-level overview of an algorithm's efficiency.
Big O notation is a powerful tool for algorithmic analysis, providing a standardized way to describe and compare the efficiency of algorithms in terms of time and space complexity as input sizes grow. It aids in making informed decisions about algorithm selection based on the scale and requirements of specific applications.

### Logarithmic Algorithm (logarithmic_algorithm):

The logarithmic algorithm uses binary search, which assumes sorted data. If you switch from numbers to words, the binary search will still work as long as the data is sorted alphabetically. If the words are not sorted or the comparison is based on a different criterion, you may need to adjust the algorithm accordingly.

```python
# Logarithmic Time Complexity (O(log n))
def logarithmic_algorithm(data):
    start_time = time.time()
    for i in range(len(data)):
        index_to_find = i * 10 #magic number 
        found_index = binary_search(data, index_to_find)
    return time.time() - start_time

def binary_search(data, target):
    low, high = 0, len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

### Quadratic Algorithm (quadratic_algorithm):

The quadratic algorithm involves a nested loop and doesn't assume anything specific about the data type. It should work similarly for both numbers and words. However, the performance impact may depend on the size and complexity of the data.

```python
# O(n)
def find_max_element(lst):
    max_element = lst[0]
    for element in lst:
        if element > max_element:
            max_element = element
    return max_element

# O(n)
def lazy_max_element(lst):
    return max(lst)

# Measure execution time for each algorithm
input_sizes = [10, 20, 30, 40, 50]
time_find_max = []
time_lazy_max = []

for size in input_sizes:
    subset_data = list(range(size))

    start_time = time.time()
    find_max_element(subset_data)
    time_find_max.append(time.time() - start_time)

    start_time = time.time()
    lazy_max_element(subset_data)
    time_lazy_max.append(time.time() - start_time)

# Plotting time complexity
plt.plot(input_sizes, time_find_max, label='find_max_element (O(n))')
plt.plot(input_sizes, time_lazy_max, label='lazy_max_element (O(n))')

print(time_find_max)
print(time_lazy_max)

plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Time Complexity Comparison')
plt.legend()
plt.show()
```

```python
import big_o


def fib_naive(n):
    if n < 0:
        return -1
    if n < 2:
        return n
    return fib_naive(n-1) + fib_naive(n-2)
print('fib_naive', big_o.big_o(fib_naive, big_o.datagen.n_, n_repeats=20, min_n=2, max_n=25)[0])

def fibonacci_memoization(n, memo={}):
    if n <= 1:
        return n
    if n not in memo:
        memo[n] = fibonacci_memoization(n - 1, memo) + fibonacci_memoization(n - 2, memo)
    return memo[n]
print("fibonacci_memoization", big_o.big_o(fibonacci_memoization, big_o.datagen.n_, n_repeats=20, min_n=2, max_n=25)[0])


# Exponential: time = -11 * 0.47^n (sec)
```

### Why Do We Care 
The results you obtained from the big_o library provide insights into the time complexity of the two Fibonacci functions you tested:

fib_naive (Naive Recursive Implementation):

* Time Complexity: Exponential, specifically O(2^n).
* The function exhibits exponential time complexity because it recursively computes Fibonacci numbers by branching into two recursive calls for each level of the recursion tree.
* The time complexity is expressed as time = 1.1E-05 * 1.6^n (sec), indicating exponential growth with the base 1.6.

fibonacci_memoization (Memoized Recursive Implementation):

* Time Complexity: Constant, specifically O(1).
* The function utilizes memoization to store previously computed Fibonacci numbers, avoiding redundant calculations.
* As a result, the time complexity becomes constant, with time = 8.7E-06 (sec), indicating that the execution time remains relatively constant regardless of the input size.

The naive recursive approach (fib_naive) suffers from exponential time complexity due to repeated and redundant calculations, resulting in a rapidly increasing execution time with larger input values. On the other hand, the memoized recursive approach (fibonacci_memoization) significantly improves the efficiency by storing and reusing previously computed values, resulting in constant time complexity and consistent performance across a range of input sizes.

```python
import numpy as np
import matplotlib.pyplot as plt
# Function definitions
def f(n):
    return 1.1E-05 * 1.6**n

def fc(n):
    return 8.7E-06

# Values for n
n_values = np.arange(0, 20, 1)  # Adjust the range as needed

# Calculate function values
f_values = f(n_values)
fc_values = np.full_like(n_values, fc(0))  # fc is constant, so it's the same for all n

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, f_values, label=r'$f = 1.1 \times 10^{-5} \times 1.6^n$', linestyle='-', marker='o')
plt.plot(n_values, fc_values, label=r'$fc = 8.7 \times 10^{-6}$', linestyle='--', marker='o')

# Additional plot configurations
plt.xlabel('n')
plt.ylabel('Function Value')
plt.title('Comparison of Functions $f$ and $fc$')
plt.legend()
plt.grid(True)
plt.show()
```

### Measuring Complexity: DataFrame Iteration in Python

Next, we will evaluate the time complexity of two distinct methods for iterating through a pandas DataFrame in Python. This analysis utilizes the big_o library to measure and compare the performance of the iterrows() and itertuples() methods.

This exploration sheds light on the algorithmic complexity associated with each method.


```python
import pandas as pd
import numpy as np
import time
import big_o

# Generate a sample DataFrame
data = {'A': np.random.randint(1, 1000, 1000),
        'B': np.random.randint(1, 1000, 1000)}

df = pd.DataFrame(data)

# Function to iterate through DataFrame using iterrows()
def iterate_df_iterrows(n):
    df_subset = df.head(n)
    for index, row in df_subset.iterrows():
        pass

# Function to iterate through DataFrame using itertuples()
def iterate_df_itertuples(n):
    df_subset = df.head(n)
    for row in df_subset.itertuples(index=False):
        pass

# Measure execution time for each iteration method
time_iterrows = big_o.big_o(iterate_df_iterrows, big_o.datagen.n_, n_repeats=20, min_n=2, max_n=1000)
time_itertuples = big_o.big_o(iterate_df_itertuples, big_o.datagen.n_, n_repeats=20, min_n=2, max_n=1000)

# Display results
print("Time complexity for iterrows():", time_iterrows[0])
print("Time complexity for itertuples():", time_itertuples[0])

# # Generate and display time complexity plots
# big_o.analyze_time(iterate_df_iterrows, n_repeats=20, min_n=2, max_n=1000).plot()
# big_o.analyze_time(iterate_df_itertuples, n_repeats=20, min_n=2, max_n=1000).plot()
```
